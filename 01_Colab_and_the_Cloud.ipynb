## 3.1 Uploading files from your local file system

- This will allow you to **upload** and **use**, say a **CSV** for example, to use as your dataset.

- **`files.upload`** returns a **dictionary** of the files which were uploaded.

- In the dictionary, **file** name is **key**, **data** uploaded is the **value**.

```python
from google.colab import files
```

```python
# Uploading files
uploaded = files.upload()
```

```python
# Checking the uploaded files
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

<a id=section302></a>

## 3.2 Downloading files to your local file system

- **`files.download`** will invoke a **browser download** of the file to the user's local computer.

```python
from google.colab import files
```

# 5. Using Interactive Parameters

<center><img src="https://raw.githubusercontent.com/insaid2018/DeepLearning/master/image/form.png" width="600" height="180" /></center>

- Using **Forms** is an **easy** way to **parameterize code**.

- From a code cell, select **`Insert â†’ Add form field`**.

- When you **change** the **value** in a **form**, the corresponding **value** in the **code** will **change**.

```python
#@title String fields

text = 'value' #@param {type:"string"}
dropdown = '1st option' #@param ["1st option", "2nd option", "3rd option"]
text_and_dropdown = 'value' #@param ["1st option", "2nd option", "3rd option"] {allow-input: true}
variable_name = "2"

print(text)
print(dropdown)
print(text_and_dropdown)
```

- You can use **`raw`** fields for **random input**.

```python
#@title Raw fields

raw_input = None #@param {type:"raw"}
raw_dropdown = raw_input #@param [1, "raw_input", "False", "'string'"] {type:"raw"}

print(raw_input)
print(raw_dropdown)
```

- Using **`date`** as a **parameter**.

```python
#@title Date fields
date_input = '2018-11-26' #@param {type:"date"}

print(date_input)
```

- Using **numerical** and **slider** inputs. 

```python
#@title Number fields
number_input = 10.0 #@param {type:"number"}
number_slider = 0 #@param {type:"slider", min:-1, max:1, step:0.1}

integer_input = 10 #@param {type:"integer"}
integer_slider = 1 #@param {type:"slider", min:0, max:100, step:1}

print(number_input)
print(number_slider)

print(integer_input)
print(integer_slider)
```

- Using **Boolean** inputs.

```python
#@title Boolean fields
boolean_checkbox = True #@param {type:"boolean"}
boolean_dropdown = True #@param ["False", "True"] {type:"raw"}

print(boolean_checkbox)
print(boolean_dropdown)
```

- Using **markdown text** as input.

```python
#@title ## Markdown
#@markdown You can also include Markdown in forms.

#@markdown ---
#@markdown ### Enter a file path:
file_path = "" #@param {type:"string"}
#@markdown ---

```
